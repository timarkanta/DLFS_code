1.Selecting the optimum k value using Silhouette Coefficient and plot the optimum k values.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
k_values = range(2, 11)
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k,n_init=10,random_state=42)
    kmeans.fit(df_cluster_data)
    score = silhouette_score(df_cluster_data, kmeans.labels_)
    silhouette_scores.append(score)

optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2

plt.plot(k_values, silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Coefficient for Optimal K')
plt.axvline(x=optimal_k, color='red', linestyle='--')
plt.show();
```

The optimum k value using Silhouette Coefficient is 9

2.Create clusters using Kmeans and Kmeans++ algorithms with optimal k value found in the
previous problem. Report performances using appropriate evaluation metrics. Compare the
results.

```python
kmeans = KMeans(n_clusters=9 ,n_init = 10,random_state=42)
kmeans.fit(df_cluster_data)
labels = kmeans.labels_

kmeans_plusplus = KMeans(n_clusters=9, init='k-means++' , n_init = 10,random_state=42)
kmeans_plusplus.fit(df_cluster_data)
labels_plusplus = kmeans_plusplus.labels_

print('KMeans Silhouette Score:', silhouette_score(df_cluster_data, labels))
print('KMeans++ Silhouette Score:', silhouette_score(df_cluster_data, labels_plusplus))
```

KMeans Silhouette Score and KMeans++ Silhouette Score are **similar**

```python
kmeans_scores = []
for i in range(50):
    kmeans = KMeans(n_clusters=9,n_init = 10,random_state=42)
    kmeans.fit(df_cluster_data)
    labels = kmeans.labels_
    score = silhouette_score(df_cluster_data, labels)
    kmeans_scores.append(score)

print('KMeans Average Silhouette Score:', sum(kmeans_scores) / len(kmeans_scores))
```

```python
kmeansplusplus_scores = []
for i in range(50):
    kmeansplusplus = KMeans(n_clusters=9,init='k-means++' , n_init = 10,random_state=42)
    kmeansplusplus.fit(df_cluster_data)
    labels = kmeansplusplus.labels_
    score = silhouette_score(df_cluster_data, labels)
    kmeansplusplus_scores.append(score)

print('KMeans Average Silhouette Score:', sum(kmeansplusplus_scores) / len(kmeansplusplus_scores))
```
